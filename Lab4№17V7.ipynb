$$\begin{cases}
3.2x_1 + 8.8x_2 = 89 \\
-14.2x_1 + 10.8x_2 = 125 \\
20.3x_1 - 7.2x_2 = 142 \\
\end{cases}$$

```python
import matplotlib.pyplot as plt
import numpy as np
import math
from matplotlib.patches import Polygon
```

```python
equations = ['$3.2x_1 + 8.8x_2 = 89$', '$-14.2x_1 + 10.8x_2 = 125$', '$20.3x_1 - 7.2x_2 = 142$']

x_1 = np.linspace(-30, 30, 1500)

f = [(89 - 3.2 * x_1)/8.8, (125 + 14.2 * x_1)/10.8, (142 - 20.3 * x_1)/-7.2]

fig, axes = plt.subplots(ncols=3, figsize=(10, 5))

for i, ax_ in enumerate(axes):
    ax_.set(xlim=(0, 25), ylim=(0, 25), title=equations[i])
    ax_.grid(linestyle=':', color='green')

axes[0].plot(x_1, f[0], '--',color='#34d2eb')
axes[1].plot(x_1, f[1], ':',color='#321652')
axes[2].plot(x_1, f[2], '-.',color='#385216')



plt.savefig('Lab4_1.png')
plt.savefig('Lab4_1.jpg')
plt.savefig('Lab4_1.svg')
```

```python
matrices = [
    np.array([[3.2, 8.8],[-14.2, 10.8]]),
    np.array([[3.2, 8.8],[20.3, -7.2]]),
    np.array([[-14.2, 10.8],[20.3, -7.2]])
]
rhs = [
    np.array([89, 125]),
    np.array([89, 142]),
    np.array([125, 142])
]

solutions = [np.linalg.solve(matrices[i], rhs[i]) for i in range(len(matrices))]
```

```python
fig = plt.figure(figsize=(16, 9), dpi=100)
ax = plt.axes()
fig.add_axes(ax)
ax.set(xlim=(-5, 30), ylim=(5, 50), xlabel='$x_1$', ylabel='$x_2$', title='All lines in one figure')
ax.grid(linestyle='-.', color='green')

ax.plot(x_1, f[0], '-.c', label='f[0]')
ax.plot(x_1, f[1], '--y', label='f[1]')
ax.plot(x_1, f[2], ':m', label='f[2]')

atans = [math.degrees(math.atan(-3.2 / 8.8)), math.degrees(math.atan(14.2 / 10.8)), math.degrees(math.atan(20.3 / 7.2))]
locations = [(1, 8.5), (9, 24), (12, 12.5)]

ax.text(*locations[0], equations[0], rotation=atans[0], rotation_mode ='anchor', transform_rotates_text = True)
ax.text(*locations[1], equations[1], rotation=atans[1], rotation_mode ='anchor', transform_rotates_text = True)
ax.text(*locations[2], equations[2], rotation=atans[2], rotation_mode ='anchor', transform_rotates_text = True)

ax.plot(*solutions[0], marker='o', color='m')
ax.plot(*solutions[1], marker='o', color='r')
ax.plot(*solutions[2], marker='o', color='c')

pts = np.array([solutions[0], solutions[1], solutions[2]])
p = Polygon(pts)
ax.add_patch(p)

ax.annotate(f"The intersection point of\n f[0] and f[1]", ha="center", xy=solutions[0], xytext=(0, 20), arrowprops=dict(arrowstyle="->"))
ax.annotate(f"The intersection point of\n f[1] and f[2]", ha="center", xy=solutions[2], xytext=(22, 25), arrowprops=dict(arrowstyle="->"))
ax.annotate(f"The intersection point of\n f[0] and f[2]", ha="center", xy=solutions[1], xytext=(15, 10), arrowprops=dict(arrowstyle="->"))

ax.legend()

plt.savefig('Lab4_2.png')
plt.savefig('Lab4_2.jpg')
plt.savefig('Lab4_2.svg')

plt.show()
```
